```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv', sep=';', decimal=',')
df.info()
```

```python
ignore_columns = [
    'Complemento',
    'Bairro',
    'Numero Rua',
    'Regiao - Sigla',
    'Estado - Sigla',
    'Municipio',
    'Nome da Rua',
    'Valor de Compra'
]
df = df.dropna()
df = df.drop(columns=ignore_columns,errors='ignore')

df['Unidade de Medida'] = df['Unidade de Medida'].str.upper().str.strip()
df['Bandeira'] = df['Bandeira'].str.upper().str.strip()
df['Produto'] = df['Produto'].str.upper().str.strip()
df.isna().sum()
```

```python
contagem_nulos = df.isnull().sum()

# Ordenar do que tem mais nulos para menos
colunas_mais_nulos = contagem_nulos.sort_values(ascending=False)

print("Colunas ordenadas por quantidade de valores nulos:")
print(colunas_mais_nulos)
```

```python
# Contar valores nulos (NaN)
contagem_nulos = df.isnull().sum()
print("Contagem de valores nulos por coluna:")
print(contagem_nulos)

# Contar valores vazios (strings vazias)
contagem_vazios = (df == '').sum()
print("\nContagem de valores vazios por coluna:")
print(contagem_vazios)

# Contar valores nulos ou vazios juntos
contagem_total = contagem_nulos + contagem_vazios
print("\nContagem total de nulos ou vazios por coluna:")
print(contagem_total)

# Contar total de linhas que possuem pelo menos um nulo ou vazio
linhas_com_vazios_ou_nulos = df[(df.isnull()) | (df == '')].any(axis=1).sum()
print(f"\nNúmero de linhas com pelo menos um valor vazio ou nulo: {linhas_com_vazios_ou_nulos}")

print("Linhas com pelo menos uma coluna nula:")
print(linhas_com_nulos)

df.head()
```
